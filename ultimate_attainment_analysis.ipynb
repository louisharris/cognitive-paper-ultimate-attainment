```python
import numpy as np
import scipy as sp
import sklearn as sk
import csv
import pandas as pd
```

```python
df = pd.read_csv('data.csv',delimiter=',')
    
    
```

```python
dfil = df.filter(items=['age','natlangs','primelangs','psychiatric','education','Eng_start','Eng_country','house_Eng','Lived_Eng_per','Eng_little','correct','elogit'])
```

filter criteria:
    at least 30 years of experience
    at most 70 years old
    
    monolinguals - age of first exposure == 0, only language is English
    immersion - simultaneous bilinguals, age of first exposure == 0, language other than English
              - later learners, 90% of life since AoFE in English speaking country (Lived_Eng_per)
    nonimmersion - at most 10% of post exposure life in English speaking country, no more than 1 year in total

```python
#filter troll responses
#dfil = dfil.mask(dfil['age'] < dfil['Eng_start'] + dfil['Eng_country_yrs'])

dfil = dfil.where(dfil['age'] <= 70)
dfil = dfil.where(dfil['Eng_start'] <= 40)
dfil = dfil.where(dfil['age'] - dfil['Eng_start'] >= 30)

#group data by above criteria
#mono = dfil.where((dfil['Eng_start'] == 0) & (dfil['natlangs'] == 'English'))
mono = dfil.where(dfil['Eng_little'] == 'monoeng')
#immers  = dfil.where((dfil['Eng_start'] == 0) & (dfil['natlangs'] != 'English') | (dfil['Lived_Eng_per'] >= 0.9))
immers = dfil.where((dfil['Eng_little'] == 'lot') | (dfil['Eng_little'] == 'bileng')) 
#nimmers = dfil.mask((dfil['Lived_Eng_per'] <= 0.1) & (dfil['Eng_country_yrs'] <= 1))
nimmers = dfil.where(dfil['Eng_little'] == 'little')

#grouped monolinguals
mono = mono.dropna(axis=0,how='any')
#grouped immersion learners
immers = immers.dropna(axis=0,how='any')
#grouped non-immersion learners
nimmers = nimmers.dropna(axis=0,how='any')

print(mono)
print(immers)
print(nimmers)

```
